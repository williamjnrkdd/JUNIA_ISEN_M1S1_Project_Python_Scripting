# Linear Regression

```python
y = data['PTS']
x1 = data['FGM']
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x)
results = results.fit()
results.summary()
```

```python
plt.scatter(x1, y)
yhat = 2.7065*x1 -0.0091
fig = plt.plot(x1,yhat, lw=4, c='orange', label='regression line')
plt.xlabel('FGM', fontsize=20)
plt.ylabel('POINTS', fontsize=20)
plt.show()
```

```python
X = data.iloc[:,8].values.reshape(-1,1)
Y = data.iloc[:, 9].values.reshape(-1,1)
X_train, X_test, y_train, y_test=train_test_split(X,Y, test_size=0.3, random_state=10)
regressor = LinearRegression()
regressor.fit(X_train, y_train)
regressor.score(X_test, y_test)
y_pred = regressor.predict(X_test)
evaluate = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
evaluate.head(10)
```
